1. 두 개의 매개변수 n, m을 전달받아 m x n개의 * 상자를 출력하는 프로그램을 함수로 작성
예: 2, 4 →  ****
             ****


```python
def box(n, m):
    for _ in range(n):
        print('*' * m)


box(2, 4)
```

2. 하나의 숫자를 전달받아 숫자의 자리 합을 구하는 함수를 작성
예: 123 → 1+2+3 = 6


```python
def sum(a):
    total = 0
    for num  in str(a):
        total += int(num)
    return total

print(sum(123)) 
```

3. 두 개의 문자열이 서로 다른 처음 위치를 반환하는 함수를 작성. 두 개의 문자열이 같으면 -1을
반환

```python
def Str(s1, s2):
    min_len = min(len(s1), len(s2))
    for i in range(min_len):
        if s1[i] != s2[i]:
            return i
    if len(s1) != len(s2):
        return min_len
    return -1

print(Str("apple", "appln"))  # 결과: 0
```

4. 숫자를 전달받아 그 수의 약수를 리스트로 반환하는 함수를 작성

```python
def number(a):
    list1 = []
    for i in range(1, 9):
        if a % i == 0:
            list1.append(i)
    return list1
    
print(number(10)) 
```

5. 문자열과 하나의 문자를 전달받아 문자열에서 문자의 위치를 모두 찾아 리스트로 반환하는 함수
를 작성

```python
def Str(s, ch):
    list1 = []
    for i in range(len(s)):
        if s[i] == ch:
            list1.append(i)
    return list1

print(Str("banana", "a"))
```
